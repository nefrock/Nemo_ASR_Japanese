```python
import os
import json
from pathlib import Path
```

```python
mode = '5x5' # 15x5 or 5x5
name = 'config-test'
```

```python
NEMO_PATH = Path('/src/NeMo/')
```

```python
template_file = NEMO_PATH / f'examples/asr/conf/config-{mode}-template.yaml'
target = NEMO_PATH / f'examples/asr/conf/{name}.yaml'
train_manifest_filepath = NEMO_PATH / 'examples/asr/conf/mix_train_manifest-test.json'
val_manifest_filepath = NEMO_PATH / 'examples/asr/conf/mix_val_manifest-test.json'
```

```python
import json
from pathlib import Path

decoder = json.JSONDecoder()
def read_json(file):
    res = []
    with open(file, 'r') as f:
        line = f.readline()
        while line:
            res.append(decoder.raw_decode(line)[0])
            line = f.readline()
    return res
```

```python
train = read_json(train_manifest_filepath)
val = read_json(val_manifest_filepath)
```

```python
texts = [m['text'] for m in train] + [m['text'] for m in val]
```

```python
texts = ''.join(texts) # 全てのテキストを一つのstrに結合
texts = list(texts) # 全てのテキストを一文字づつのstrのリストに変換
chars = sorted(list(set(texts))) # 重複のない文字のリストを取得
```

```python
num_classes = len(chars) # 3227
```

```python
labels = str(chars)
labels = labels.replace('\'', '') # 普通の文字はクオートを除く
```

```python
labels = labels.replace('[ ', '[\' \' ') # 空文字はクオートしなければならない
```

```python
print(labels)
```

```python
with open(template_file, 'r') as f:
    template = f.read()
```

```python
template = template.replace('LABELS', labels)
template = template.replace('NUM_CLASSES', str(num_classes))
template = template.replace(f'QuartzNet{mode}', f'QuartzNet{mode}-{name}')
template = template.replace('TRAIN_MANIFEST_FILEPATH', str(train_manifest_filepath.absolute()))
template = template.replace('VAL_MANIFEST_FILEPATH', str(val_manifest_filepath.absolute()))
```

```python
with open(target, 'w') as f:
    f.write(template)
```
